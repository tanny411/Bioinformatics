```
def BWT(text):
  bwt = []
  for i,_ in enumerate(text):
    bwt.append(text[i:]+text[:i])
  bwt = sorted(bwt)
  for i in bwt:
    print(i)
  return ''.join(v[-1] for v in bwt)
```

```
def num(text):
    a,c,t,g=0,0,0,0
    for j,i in enumerate(text):
      if i=='A':
        a+=1
        text[j]='A'+str(a)
      elif i=='G':
        g+=1
        text[j]='G'+str(g)
      elif i=='C':
        c+=1
        text[j]='C'+str(c)
      elif i=='T':
        t+=1
        text[j]='T'+str(t)
    return text
```

```
def invBWT(text):
  text = list(text)
  srt = sorted(text)
  text = num(text)
  srt = num(srt)
  # print(text,srt)
  ans = ['$']
  # print(type(ans))
  for i in range(len(srt)-1):
    idx = srt.index(ans[-1])
    ans.append(text[idx])
  return ''.join([x[0] for x in ans][::-1])
```

## Task 5

```
def appx(text, ls, x):

  def ham(a,b):
    diff=0
    for i,j in zip(a,b):
      if i!=j:
        diff+=1
    return diff

  def khujo(text,pat,mis):
    ln = len(pat)
    ret=[]
    for i in range(len(text)-ln+1):
      if ham(text[i:i+ln],pat) <= mis:
        ret.append(i)
    return ret

  ret=[]
  for pat in ls:
    ret+=khujo(text,pat,x)
  return sorted(ret)

text = "ACATGCTACTTT"
pats = "ATT GCC GCTA TATT".split()
mis = 1
appx(text,pats,mis)
```

## Task 6

```
def appx(text, ls):

  def khujo(text,pat):
    ln = len(pat)
    ret=[]
    for i in range(len(text)-ln+1):
      if text[i:i+ln] == pat:
        ret.append(i)
    return ret

  ret=[]
  for pat in ls:
    ret+=khujo(text,pat)
  return sorted(ret)

text = "AATCGGGTTCAATCGGGGT"
pats = "ATCG GGGT".split()
appx(text,pats)
```
