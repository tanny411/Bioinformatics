```
from collections import defaultdict 

def clump(genome,k,L,t):
  mp=defaultdict(list)
  for i in range(len(genome)-k):
    pat=genome[i:i+k]
    if pat not in mp.keys():

      #find all patters
      for j in range(i,len(genome)-k):
        if genome[j:j+k]==pat:
          mp[pat].append(j)
  
#   print(mp)
  
  ans = set()
  
  for key,val in mp.items():
    if len(val)>=t:
      for ix in range(len(val)-t+1):
        if val[ix+t-1]+k-val[ix]<=L:
          ans.add(key)
          break;
          
#   print(ans)
  for i in ans:
    print(i,end=' ')
```

```
import itertools

def medianString(k,dnas):
  def ham(a,b):
    ret=0;
    for i in range(len(a)):
      if a[i]!=b[i]:
        ret+=1
    return ret
  
  def d(pat,dna):
    mn=len(dna)
    for i in range(len(dna)-k):
      mn=min(mn,ham(pat,dna[i:i+k]))
    return mn
  
  def d_all(pat,dnas):
    ret=0;
    for dna in dnas:
      ret+=d(pat,dna)
    return ret
  
  kmers=[''.join(i) for i in itertools.product('ATCG', repeat = k)]
  
  ret=""
  mn=len(dnas[0])
  for kmer in kmers:
    m = d_all(kmer,dnas)
    if m<mn:
      mn=m;
      ret=kmer
    
  return ret
```

```
def Profile_most_Probable_k_mer(s,k,mat):
  # mat = 4 by k size
  
  mp={'A':0,'C':1,'G':2,'T':3}
  
  def prob(pat):
    val = 1.0
    for ix,i in enumerate(pat):
#       print(type(mp[i]),type(ix))
      val*=mat[mp[i],ix]
    return val
  
  mx=0
  ans=""
  for i in range(len(s)-k):
    m = prob(s[i:i+k])
    if m>mx:
      mx=m
      ans=s[i:i+k]
  return ans;
```
