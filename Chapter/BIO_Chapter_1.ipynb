## BA1E	Find Patterns Forming Clumps in a String

```
from collections import defaultdict 

def clump(genome,k,L,t):
  mp=defaultdict(list)
  for i in range(len(genome)-k):
    pat=genome[i:i+k]
    if pat not in mp.keys():

      #find all patters
      for j in range(i,len(genome)-k):
        if genome[j:j+k]==pat:
          mp[pat].append(j)
  
#   print(mp)
  
  ans = set()
  
  for key,val in mp.items():
    if len(val)>=t:
      for ix in range(len(val)-t+1):
        if val[ix+t-1]+k-val[ix]<=L:
          ans.add(key)
          break;
          
#   print(ans)
  for i in ans:
    print(i,end=' ')
```

## BA1H Find All Approximate Occurrences of a Pattern in a String

```
def appPatternMatch(pat,text,d):
  '''
  We say that a  Pattern appears as a substring of Text with at most d mismatches if there is some 
  substring Pattern of Text having d or fewer mismatches with Pattern
  '''
  ln = len(pat)
  ans=[]
  
  for i in range(len(text)-ln):
    if hamming(pat,text[i:i+ln])<=d:
      ans.append(i)
  return ans

appPatternMatch('ATTCTGGA','CGCCCGAATCCAGAACGCATTCCCATATTTCGGGACCACTGGCCTCCACGGTACGGACGTCAATCAAATGCCTAGCGGCTTGTGGTTTCTCCTACGCTCC',3)
```

```
def PatternToNumber(pat):
  pats = [''.join(i) for i in itertools.product('ACGT', repeat = len(pat))]
  return pats.index(pat)

PatternToNumber('AGT')
```

## BA1M Implement NumberToPattern

```
def NumberToPattern(index,k):
  pats = [''.join(i) for i in itertools.product('ACGT', repeat = k)]
  return pats[index]

NumberToPattern(45,4)
```

## BA1N Generate the d-Neighborhood of a String

```
import itertools
```

```
def dNeighbour(dna,d):
  pats = [''.join(i) for i in itertools.product('ACGT', repeat = len(dna))]
  ret=[]
  for pat in pats:
    if hamming(dna,pat)<=d:
      ret.append(pat)

  return ret

dNeighbour("ACG",1)
```
